## Aufgabe 2.1 -Integrator backstepping- asymptotic control


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the system of differential equations
#gain 
k=1
k2=0.1
def system(state, t, k1,k2):
    x1, x2 = state
    u=(-k-3*x1**2)*(x1**3+x2)-x1-k2*(x2-(-k*x1-x1**3))
    dx1dt = x1**3 + x2
    dx2dt = u
    return [dx1dt, dx2dt]

# Initial conditions
x1_0 = 1.0  # Example initial condition for x1
x2_0 = 4  # Example initial condition for x2
initial_state = [x1_0, x2_0]

#time 
t = np.linspace(0, 10, 100)



# Solve the system 
solution = odeint(system, initial_state, t, args=(k,k2))

# Extract x1 and x2 from solution
x1 = solution[:, 0]
x2 = solution[:, 1]

# Plot the results
plt.figure()
plt.plot(t, x1, label='x1(t)')
plt.plot(t, x2, label='x2(t)')
plt.xlabel('Time t')
plt.ylabel('States x1, x2')
plt.legend()
plt.title('solutions')
plt.show()
```
