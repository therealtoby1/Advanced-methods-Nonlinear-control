## Aufgabe 2.1 -Integrator backstepping- asymptotic control


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the system 



def system(state, t):
    x=state
    u= (-x**2 -x**4)/(x**3+np.sin(x)+1.5) -x*(x**2+np.sin(x)+1.5)
    d= t*x #randomly chosen function (could also be a constant)
    dxdt = x**3 + ( x**2 + np.sin(x) + 1.5)*(u+d)
    
    return dxdt

# Initial conditions
x_0 = 1  # Example initial condition for x


#time 
t = np.linspace(0, 10, 100)



# Solve the system 
solution = odeint(system,y0=x_0,t=t)

# Extract x1 and x2 from solution
x = solution[:]


# Plot the results
plt.figure()
plt.plot(t, x, label='x1(t)')
plt.xlabel('Time t')
plt.ylabel('State x')
plt.legend()
plt.title('solutions')
plt.show()

```
